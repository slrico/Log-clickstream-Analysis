# Advanced Pre processing

Handle Missing Values
- Identify Missing Values: Determine the extent of missing values in each column. Use functions like isnull().sum() in pandas to detect missing values. Imputation or Removal: Decide whether to fill missing values with a mode/mean/median or to drop rows/columns based on the amount of missing data: For categorical columns, consider using the mode. For numerical columns, consider mean or median depending on distribution.

```python
import pandas as pd

def preprocess_data(df):
    """
    Preprocess the DataFrame to handle missing values.

    Parameters:
    - df (pd.DataFrame): The DataFrame to preprocess.

    Returns:
    - df (pd.DataFrame): The preprocessed DataFrame.
    """

    # Step 1: Replace '?' with NaN
    df.replace(['?', ' ?', ' ? '], pd.NA, inplace=True)

    # Optional: Strip whitespace from all string entries in the DataFrame
    df = df.apply(lambda x: x.str.strip() if x.dtype == "object" else x)

    # Step 2: Print missing values before preprocessing
    print("\nMissing values before preprocessing:")
    print(df.isnull().sum())

    # Step 3: Impute or Remove Missing Values
    for column in df.columns:
        if df[column].isnull().any():  # Check if there are any missing values
            if df[column].dtype == 'object':  # Categorical data
                mode_value = df[column].mode()[0]
                df[column].fillna(mode_value, inplace=True)
            else:  # Numerical data
                if df[column].skew() > 1 or df[column].skew() < -1:  # Check for skewness
                    median_value = df[column].median()
                    df[column].fillna(median_value, inplace=True)
                else:
                    mean_value = df[column].mean()
                    df[column].fillna(mean_value, inplace=True)

    # Step 4: Print missing values after preprocessing
    print("\nMissing values after preprocessing:")
    print(df.isnull().sum())

    return df  # Return the cleaned DataFrame

if __name__ == "__main__":
    file_name = 'incident_event_log.csv'  # Use the same file as before
    df = read_and_display_csv(file_name)  # Read the CSV file directly

    if df is not None:  # Ensure df was successfully created
        df = preprocess_data(df)  # Preprocess the DataFrame

        # At this point, df contains the cleaned data and overwrites the original DataFrame
```

Temporal Convolutional Networks (TCNs)
Compared to RNNs, Temporal Convolutional Networks (TCNs) can effectively handle sequences while maintaining more parallelism, making them efficient and often more powerful. TCNs use causal convolutions, ensuring the model only uses past information to make predictions.

```python
import torch
import torch.nn as nn
import torch.nn.functional as F

class TCN(nn.Module):
    def __init__(self, input_size, num_channels):
        super(TCN, self).__init__()
        layers = []
        for i in range(len(num_channels)):
            dilation = 2 ** i
            layers.append(nn.Conv1d(in_channels=input_size if i == 0 else num_channels[i - 1],
                                    out_channels=num_channels[i],
                                    kernel_size=3,
                                    dilation=dilation,
                                    padding=dilation))
            layers.append(nn.BatchNorm1d(num_channels[i]))
            layers.append(nn.ReLU())
        self.network = nn.Sequential(*layers)

    def forward(self, x):
        x = self.network(x)
        return x
```

```python
import torch
import torch.nn as nn
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Load and prepare your dataset
df = pd.read_csv('log_data.csv', delimiter=';')
# Preprocess your features, e.g., scale numerical values, encode categorical ones, etc.
# Example: Assume 'resolution_time' is a target we want to predict
df['resolved_at'] = pd.to_datetime(df['resolved_at'])
df['opened_at'] = pd.to_datetime(df['opened_at'])
df['resolution_time'] = (df['resolved_at'] - df['opened_at']).dt.total_seconds() / 3600  # In hours

# Selecting features and splitting the dataset
features = df[['reassignment_count', 'reopen_count', 'sys_mod_count']]  # Add other relevant features
target = df['resolution_time']

# Split into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(features.values, target.values, test_size=0.2, random_state=42)

# Scale the features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Define a Transformer model for regression
class TransformerModel(nn.Module):
    def __init__(self, input_size, num_heads, num_layers, output_size):
        super(TransformerModel, self).__init__()
        self.input_size = input_size
        self.mha = nn.MultiheadAttention(embed_dim=input_size, num_heads=num_heads)
        self.fc1 = nn.Linear(input_size, 128)
        self.fc2 = nn.Linear(128, output_size)
        self.layers = nn.ModuleList([nn.TransformerEncoderLayer(d_model=input_size, nhead=num_heads) for _ in range(num_layers)])

    def forward(self, x):
        x = x.unsqueeze(1)  # Add sequence dimension
        for layer in self.layers:
            x = layer(x)
        x = x.mean(dim=1)  # Global average pooling
        x = F.relu(self.fc1(x))
        return self.fc2(x)

# Model initialization
input_size = X_train_scaled.shape[1]
model = TransformerModel(input_size=input_size, num_heads=4, num_layers=2, output_size=1)

# Loss and optimizer
criterion = nn.MSELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)

# Training loop
for epoch in range(100):  # Number of epochs
    model.train()
    optimizer.zero_grad()
    X_batch = torch.tensor(X_train_scaled, dtype=torch.float32)
    y_batch = torch.tensor(y_train, dtype=torch.float32).unsqueeze(-1)

    predictions = model(X_batch)
    loss = criterion(predictions, y_batch)
    loss.backward()
    optimizer.step()

    if (epoch + 1) % 10 == 0:
        print(f'Epoch [{epoch + 1}/100], Loss: {loss.item():.4f}')

# Evaluate the model
model.eval()
with torch.no_grad():
    X_test_tensor = torch.tensor(X_test_scaled, dtype=torch.float32)
    y_test_tensor = torch.tensor(y_test, dtype=torch.float32).unsqueeze(-1)
    predictions = model(X_test_tensor)
    test_loss = criterion(predictions, y_test_tensor)
    print(f'Test Loss: {test_loss.item():.4f}')
```
